# 1. Немного про numpy

До этого мы с вами работали с пакетом `pandas`. Он помогал нам обрабатывать таблички. У него есть брат, пакет `numpy`. Он тоже позволяет работать с данными, но в отличие от `pandas`, он смотрит на них, как на матрицы. Также пакет содержит в себе кучу полезного функционала для разной математики. 

```python
import numpy as np
```

По большей части пакет позволяет работать с массивами. Вся работа аналогична тому, что мы делали с листами. Например, вот так можно создать массив и сделать срез в нём: 

```python
x = np.array([1,-10,3,0,1,75,3])
x[3:6]
```

Но есть и разница. Например, если сложить два листа, получтся огромный лист:

```python
x = [1,-10,3,0,1,75,3]
y = [9,1,567,1,4557,1123,1]
x + y
```

А если сложить два нумпаевских массива, то все числа сложатся поэлементно. 

```python
x = np.array([1,-10,3,0,1,75,3])
y = np.array([9,1,567,1,4557,1123,1])
x + y
```

Если лист умножить на три, он повторится три раза.

```python
x = [1, 2, 3]
3 * x
```

Если нумпаевский массив умножить на три, все элементы внутри него умножатся на три.

```python
x = np.array([1, 2, 3])
3 * x
```

Добавим несколько столбцов, которые будут отражать _возраст видео_ в более удобной форме. Заодно научимся работать с датами. 

```python
from datetime import datetime  # пакет для работы с датами
```

Сейчас все даты записаны как текст. Буква $T$ отделяет дату от времени, буква $Z$ означает, что это время по нулевому мередиану (Григвич).

```python
x = youtube.publishedAt.iloc[0]
x
```

Распарсим дату пакетом `datetime`, чтобы питон понимал сколько часов, дней минут и т.д. было в момент публикации видео.

```python
y = datetime.strptime(x, "%Y-%m-%dT%H:%M:%S.%fZ")
y
```

Можно представить время в целого числа: 

```python
y.timestamp()  # число секунд с 1 января 1970 года до публикации видео
```

Подробнее про подобный формат времени [читай тут.](https://ru.wikipedia.org/wiki/Unix-время) Кстати говоря, с таким форматом записи времени связана [проблема 2038 года.](https://ru.wikipedia.org/wiki/Проблема_2038_года) Именно тогда может настать цифровой апокалипсис. Почитайте про это, чтобы быть в курсе. 

Между распаршеными датами можно считать число секунд/минут/часов и тп 

```python
date0 = datetime.strptime(youtube.publishedAt.iloc[0], "%Y-%m-%dT%H:%M:%S.%fZ")
date1 = datetime.strptime(youtube.publishedAt.iloc[1], "%Y-%m-%dT%H:%M:%S.%fZ")

diff = date0 - date1
diff
```

Завернём это всё в функции. 

```python
def get_ts(timestamp_iso8601):
    timestamp = datetime.strptime(timestamp_iso8601, "%Y-%m-%dT%H:%M:%S.%fZ")
    return timestamp.timestamp()

def get_age(published_at, now = "2019-03-15T00:00:00.000Z"):
    age = datetime.strptime(now, "%Y-%m-%dT%H:%M:%S.%fZ") -  datetime.strptime(published_at, "%Y-%m-%dT%H:%M:%S.%fZ")
    return int(age.total_seconds() / 86400)
```

```python
# Ещё не забыли, как работает apply? 
youtube['publishedAt_ts'] = youtube.publishedAt.apply(get_ts)
youtube['video_age'] = youtube.publishedAt.apply(get_age)
```

## 3.2  Усатый ящик  и seaborn 

Многие картинки в matplotlib рисовать не очень удобно и приходится делать много циклов. На пример такого кода вы можете посмотреть в конце тетрадки. Для простоты был сделан специальный пакет для графики `seaborn`. В нём все картинки строятся в одну-две команды.

Давайте построим в seaborn boxplot.  Некоторые называют его "ящик с усиками". 

![](https://upload.wikimedia.org/wikipedia/commons/3/32/Densityvsbox.png)

Представляет из себя визуальное описание одномерной величины: отражает медиану, верхнюю и нижнюю квартиль, выбросы и минимальное и максимальное значения.

# 4. Более сложные картинки

Некоторые картинки приходится строить по циклу. Давайте посмотрим как это делается, чтобы в будущем у нас не было с этим проблем. 

```python
cat_features = ['definition', 'license', 'music_style']
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharey=True)

for i in range(len(cat_features)):
    sns.countplot(x=cat_features[i], data=youtube, ax=axes[i // 2, i % 2])
```

Ещё пример! Наносим каждую точку со своим цветом в зависимости от жанра. Размер точки - как часто жанр встречается! 

```python
sizes = df_log.groupby('music_style').agg('count')['dislikeCount']
sizes
```

```python
sizes = sizes.to_dict()
sizes
```

```python
# При желании можно импортировать цветовую схему:  http://seaborn.pydata.org/tutorial/color_palettes.html
colors = 'bgrcmyk'

colors = dict(zip(df_log.music_style.unique(), list(colors)))
colors
```

```python
styles = list(sizes.keys())

for i,y in enumerate(styles):
    plt.scatter(i, 1, c=colors[y], alpha=0.5, label=str(y), s=sizes[y])
    
plt.xticks(ticks=list(range(len(styles))) , labels=styles, size=18);
```
